# GRADIENTE DESCENDIENTE

```python
def costo(x, y, w):
  # Función costo inicial f = w0 + w1 x
  f_wb = w[0] * x + w[1]
  # Función costo del algoritmo (f - y) ^ 2
  cost = np.sum((f_wb - y) ** 2)
  cost /= (2 * len(x))

  return cost
```

```python
def derivada_costo(x, y, w):
  dw = np.zeros(len(w))
  fw = 0
  for i in range(len(w)):

    fw += w[i] * (x ** i)

  for j in range(len(w)):
    #Esto es la derivada para obtener w0
    dw[j] = np.sum((fw - y) * (x** j))/ len(x)

  return dw
```

```python
import math as math

def gradiente_descendente(x, y, w, alpha, num_iter, funcion_costo, funcion_gradiente):
  for i in range(num_iter):
    dj_dw = funcion_gradiente(x, y, w)
    for j in range(len(w)):
      w[j] = w[j] - alpha * dj_dw[j]
    cost = funcion_costo(x, y, w)
    # Print cost every intervals 10 times or as many iter if < 10
    if i % math.ceil(num_iter / 10) == 0:
      print(f"Iteration {i:4}: Cost {cost: 0.5e} ",
            f"w[0]: {w[0]: 0.5e}, w[1]: {w[1]: 0.5e}")
  return w, costo
```

```python
# Parametros iniciales
w0, w1 = 11, -1
# Gradient descent settings
iterations = 50000
tmp_alpha = 1e-2
# run gradient
w_final, costo = gradiente_descendente(x_norm, y, [w0, w1], tmp_alpha,
                                iterations, costo, derivada_costo)

print(f"W econtradas: {w_final}")
```

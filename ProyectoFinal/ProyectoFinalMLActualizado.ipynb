Se creo una función para cargar varias imagenes de los archivos en drive

```python
# Función para cargar datos de manera más efectiva
def load_data(data_directory):
    images = []  # Lista para almacenar las imágenes
    labels = []  # Lista para almacenar las etiquetas
    classes = []  # Lista para almacenar los nombres de las clases
    label_encoder = LabelEncoder()  # Inicializar el codificador de etiquetas

    # Iterar sobre las clases dentro del directorio de datos
    for class_name in os.listdir(data_directory):
        class_path = os.path.join(data_directory, class_name)
        classes.append(class_name)  # Agregar el nombre de la clase a la lista de clases

        # Iterar sobre los archivos de imágenes dentro de cada clase
        for filename in os.listdir(class_path):
            filepath = os.path.join(class_path, filename)
            img = cv2.imread(filepath)  # Leer la imagen utilizando OpenCV

            # Verificar si la imagen se ha cargado correctamente
            if img is not None:
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convertir la imagen a RGB
                img = cv2.resize(img, (224, 224))  # Redimensionar la imagen a 224x224 píxeles
                img = img / 255.0  # Normalizar los valores de píxeles para que estén en el rango [0, 1]
                images.append(img)  # Agregar la imagen a la lista de imágenes
                labels.append(class_name)  # Agregar la etiqueta correspondiente a la lista de etiquetas

    # Convertir las listas de imágenes y etiquetas a arrays numpy
    images = np.array(images)
    labels = label_encoder.fit_transform(labels)  # Codificar las etiquetas numéricamente

    # Devolver las imágenes, etiquetas codificadas y nombres de las clases
    return images, labels, classes
```

## Definir conjunto de entrenamiento y prueba

Los datos usados venían ya clasificados como Train y Test, entonces se definió como ya venían estructurados en el dataset

```python
# Define las rutas de los directorios
train_directory = '/content/drive/MyDrive/ML2023/jellyfish/Train_Test_Valid/Train'
test_directory = '/content/drive/MyDrive/ML2023/jellyfish/Train_Test_Valid/test'

# Carga los datos de entrenamiento y prueba
train_images, train_labels, classes = load_data(train_directory)
test_images, test_labels, _ = load_data(test_directory)

# División de los conjuntos de entrenamiento y prueba
train_images, val_images, train_labels, val_labels = train_test_split(train_images, train_labels, test_size=0.2, random_state=42)

# Guardar los conjuntos de entrenamiento, validación y prueba en archivos npy
np.save('/content/drive/MyDrive/ML2023/jellyfish/train_images.npy', train_images)
np.save('/content/drive/MyDrive/ML2023/jellyfish/train_labels.npy', train_labels)
np.save('/content/drive/MyDrive/ML2023/jellyfish/val_images.npy', val_images)
np.save('/content/drive/MyDrive/ML2023/jellyfish/val_labels.npy', val_labels)
np.save('/content/drive/MyDrive/ML2023/jellyfish/test_images.npy', test_images)
np.save('/content/drive/MyDrive/ML2023/jellyfish/test_labels.npy', test_labels)

# Guardar las clases en un archivo txt para referencia
with open('/content/drive/MyDrive/ML2023/jellyfish/classes.txt', 'w') as file:
    for class_name in classes:
        file.write(class_name + '\n')
```

## Entrenamiento y Evaluación del Modelo
El modelo se entrena con los datos de entrenamiento y validación. Se compila con la función de pérdida 'sparse_categorical_crossentropy' y se utiliza el optimizador 'adam'.



### Arquitectura de la red neuronal convolucional:

* Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)):
32 filtros de tamaño 3x3 con función de activación ReLU.
Tamaño de entrada de la imagen: 224x224 píxeles con 3 canales de color (RGB).
Las capas convolucionales extraen características de las imágenes mediante convoluciones, detectando patrones locales.

* Conv2D(64, (3, 3), activation='relu'):
Capa convolucional con 64 filtros de tamaño 3x3 y activación ReLU.
Aumentar la profundidad de los filtros puede permitir la detección de patrones más complejos en las imágenes.

* Conv2D(128, (3, 3), activation='relu'):
Otra capa convolucional con 128 filtros y activación ReLU.
A medida que avanzamos en la red, se busca capturar características más abstractas.

* Conv2D(256, (3, 3), activation='relu'):
Capa convolucional con 256 filtros y activación ReLU.
Aumentar aún más la profundidad para obtener representaciones más complejas.

* MaxPooling2D((2, 2)):
Operación de agrupación con ventanas de tamaño 2x2.
Reduce la dimensionalidad de las características extraídas, conservando la información más relevante.
* Capas Densas (Dense):
Dense(512, activation='relu'):
Capa densa con 512 neuronas y activación ReLU.
Las capas densas procesan las características extraídas y realizan la clasificación final.

* Dropout(0.5):
Regularización para prevenir el sobreajuste (overfitting).
Desactiva aleatoriamente el 50% de las neuronas durante el entrenamiento para mejorar la generalización del modelo, probando con el modelo utilizar Dropout aumentaba radicalmente la precisión.

* Dense(len(classes), activation='softmax'):
Capa densa de salida con un número de neuronas igual al número de clases en el conjunto de datos y activación softmax.

En general, la arquitectura está diseñada para capturar jerarquías de características en las imágenes, desde detalles simples hasta patrones complejos.
El uso de capas de convolución y agrupación permite aprender representaciones cada vez más abstractas de las imágenes a medida que se avanza en la red.
La regularización con Dropout ayuda a mitigar el sobreajuste al introducir variabilidad durante el entrenamiento.
Las funciones de activación ReLU en las capas convolucionales y densas permiten la no linealidad, ayudando a la red a aprender relaciones complejas entre las características.

```python
# Definir la arquitectura de la red neuronal
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(256, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(512, activation='relu'),
    Dropout(0.5),
    Dense(len(classes), activation='softmax')
])

# Compilar modelo
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Definición de los hiperparámetros
batch_size = 32
epochs = 9

# Entrenamiento del modelo
history = model.fit(train_images, train_labels, batch_size=batch_size, epochs=epochs,
                    validation_data=(val_images, val_labels))

# Gráfico de la precisión y pérdida del modelo a lo largo de las épocas
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Evaluar el modelo en el conjunto de prueba
loss, accuracy = model.evaluate(val_images, val_labels)
print(f"Pérdida en conjunto de prueba: {loss}")
print(f"Precisión en conjunto de prueba: {accuracy * 100} %")
```

## Ponemos a prueba el modelo

```python
# Predicciones del modelo en el conjunto de prueba
predictions = model.predict(test_images)

# Mostrar 10 imágenes aleatorias junto con las predicciones del modelo
plt.figure(figsize=(15, 8))
for i, idx in enumerate(random.sample(range(len(test_images)), 10), 1):
    plt.subplot(2, 5, i)
    plt.imshow(test_images[idx])
    predicted_class = classes[np.argmax(predictions[idx])]
    true_class = classes[test_labels[idx]]
    plt.title(f'Pred: {predicted_class}\nTrue: {true_class}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

## Pruebas con otras arquitecturas:



```python
def create_model(input_shape, num_classes, architecture='default'):
    model = Sequential()  # Inicializar el modelo en cada caso
    # Menos capas y con un dropout más pequeño
    if architecture == 'smaller':
        model.add(Conv2D(16, (3, 3), activation='relu', input_shape=input_shape))
        model.add(MaxPooling2D((2, 2)))
        model.add(Flatten())
        model.add(Dense(64, activation='relu'))
        model.add(Dropout(0.3))
        model.add(Dense(num_classes, activation='softmax'))
    # Mismas capas pero con menor número de neuronas, mismo dropout.
    elif architecture == 'larger':
        model.add(Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
        model.add(MaxPooling2D((2, 2)))
        model.add(Conv2D(64, (3, 3), activation='relu'))
        model.add(MaxPooling2D((2, 2)))
        model.add(Conv2D(128, (3, 3), activation='relu'))
        model.add(MaxPooling2D((2, 2)))
        model.add(Flatten())
        model.add(Dense(256, activation='relu'))
        model.add(Dropout(0.5))
        model.add(Dense(num_classes, activation='softmax'))
    else:
        raise ValueError("Invalid architecture type. Choose from 'smaller', or 'larger'.")

    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
# Ahora vamos a probar las diferentes arquitecturas con tus datos
architectures_to_try = ['smaller', 'larger']
```

```python
for architecture in architectures_to_try:
    # Definir la forma de entrada y el número de clases
    image_shape = train_images[0].shape
    num_classes = len(classes)

    # Crear el modelo
    new_model = create_model(input_shape=image_shape, num_classes=num_classes, architecture=architecture)

    # Entrenar el modelo con tus datos
    epochs = 15
    history = new_model.fit(train_images, train_labels, epochs=epochs, validation_data=(val_images, val_labels))

    # Evaluar el modelo con los datos de prueba
    loss, accuracy = new_model.evaluate(test_images, test_labels)
    print(f"Architecture: {architecture}")
    print(f"Pérdida en conjunto de prueba: {loss}")
    print(f"Precisión en conjunto de prueba: {accuracy * 100} %")
```

```python
# Gráfico de la precisión y pérdida del modelo a lo largo de las épocas
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()

plt.tight_layout()
plt.show()
```
